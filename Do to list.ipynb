```python
from tkinter import *
from tkinter import messagebox
 
import sqlite3 as sql

def add_task():  
    task_string = task_field.get()  
    if len(task_string) == 0:  
        messagebox.showinfo('Error', 'Field is Empty.')  
    else:  
        tasks.append(task_string)   
        the_cursor.execute('insert into tasks values (?)', (task_string ,))    
        list_update()  
        task_field.delete(0, 'end')  
  
  
def list_update():  
    clear_list()  
    for task in tasks:    
        task_listbox.insert('end', task)  



def delete_task():  
    try:  
        the_value = task_listbox.get(task_listbox.curselection())   
        if the_value in tasks:  
            tasks.remove(the_value)   
            list_update()  
            the_cursor.execute('delete from tasks where title = ?', (the_value,))  
    except:   
        messagebox.showinfo('Error', 'No Task Selected. Cannot Delete.')   
        
def update_task():
    try:
        selected_task = task_listbox.get(task_listbox.curselection())
        if selected_task in tasks:
            tasks.remove(selected_task)
            list_update()  
            the_cursor.execute('delete from tasks where title = ?', (the_value,))
            tasks.append(task_string)   
        the_cursor.execute('insert into tasks values (?)', (task_string ,))    
        list_update()
    except:
        messagebox.showinfo('Error', 'No Task Selected. Cannot Update.')
  
 
def clear_list():   
    task_listbox.delete(0, 'end')  
  
 
def close():  
    print(tasks)  
    guiWindow.destroy()  
    
  
# main function  
if __name__ == "__main__":  
    guiWindow = Tk()    
    guiWindow.title("To-Do List ")  
    guiWindow.geometry("665x400+550+250")  
    guiWindow.configure(bg = "light gray")  
  
    the_connection = sql.connect('listOfTasks.db')    
    the_cursor = the_connection.cursor()  
    the_cursor.execute('create table if not exists tasks (title text)')  
  
  
    tasks = []  
    functions_frame = Frame(guiWindow, bg = "light gray") 
    functions_frame.pack(side = "top", expand = True, fill = "both")  
    task_label = Label( functions_frame,text = "Enter the Task:",  
        background = "light gray", 
        foreground="black"
    )  
    task_label.place(x = 20, y = 30)  
    
    task_field = Entry(  
        functions_frame,    
        width = 42,  
        foreground="Black",
        background = "White",  
    )  
    task_field.place(x = 180, y = 30)  
  
    add_button =Button(  
        functions_frame,  
        text = "Add Task",  
        width = 15,
        bg='light gray',
        command = add_task,
        
    )  
    del_button = Button(  
        functions_frame,  
        text = "Delete Task",  
        width = 15,
        bg='light gray',
        command = delete_task,  
    )  
    update_button =Button(  
        functions_frame,  
        text = "Update Task",  
        width = 15,
        bg='light gray',
        command = add_task,
        
    )  
    
    exit_button = Button(  
        functions_frame,  
        text = "Exit",  
        width = 52,
        bg='light gray',
        command = close  
    )  
    
    add_button.place(x = 120, y = 80,)
    update_button.place(x= 270, y= 80)
    del_button.place(x = 430, y = 80)   
    exit_button.place(x = 150, y = 330)  
  
   
    task_listbox = Listbox(  
        functions_frame,  
        width = 63,  
        height = 7,  
        font="bold",  
        background = "WHITE",
        foreground="black",    
        
    )  
    
    task_listbox.place(x = 50, y = 140)  
  
      
     
    list_update()  
    guiWindow.mainloop()  
   
    the_connection.commit()  
    the_cursor.close()
```
